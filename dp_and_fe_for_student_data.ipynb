```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder
import re
```

ดึงข้อมูลจาก dataset

```python
df =pd.read_csv('student.csv')
```

แสดงตัวอย่างข้อมูลก่อนทำความสะอาด

```python
df
```

copy จาก dataset หลักมาไว้ใน df_clean

```python
df_clean = df.copy()
```

ลบช่องว่างและทำให้เป็นมาตรฐาน

```python
df_clean.columns = df_clean.columns.str.strip()
```

แก้ไขค่า GPA ที่ผิดปกติโดยกำหนดให้ GPA มากกว่า 0 แต่น้อยกว่า 4

```python
df_clean.loc[df_clean['GPA'] < 0, 'GPA'] = np.nan  # Remove negative GPA
df_clean.loc[df_clean['GPA'] > 4.0, 'GPA'] = 4.0   # Cap at 4.0
```

แก้ไขค่า อายุ ที่ผิดปกติโดยกำหนดให้ อายุ นั้นมากกว่า 17 แต่น้อยกว่า 30

```python
df_clean.loc[df_clean['อายุ'] > 30, 'อายุ'] = np.nan  # Remove unrealistic age (150)
df_clean.loc[df_clean['อายุ'] < 17, 'อายุ'] = np.nan
```

แก้ไขค่า ส่วนสูง ที่ผิดปกติโดยกำหนดให้ ส่วนสูง นั้นมากกว่า 140 แต่น้อยกว่า 200

```python
df_clean.loc[df_clean['ส่วนสูง_cm'] < 140, 'ส่วนสูง_cm'] = np.nan
df_clean.loc[df_clean['ส่วนสูง_cm'] > 200, 'ส่วนสูง_cm'] = np.nan
```

แก้ไขค่า น้ำหนัก ที่ผิดปกติโดยกำหนดให้ น้ำหนัก นั้นมากกว่า 35 แต่น้อยกว่า 120

```python
df_clean.loc[df_clean['น้ำหนัก_kg'] < 35, 'น้ำหนัก_kg'] = np.nan
df_clean.loc[df_clean['น้ำหนัก_kg'] > 120, 'น้ำหนัก_kg'] = np.nan
```

แก้ไขค่า ชั่วโมงเรียน ต่อสัปดาห์ ที่ผิดปกติโดยกำหนดให้ ชั่วโมงเรียน ต่อสัปดาห์ นั้นมากกว่า 5 แต่น้อยกว่า 40

```python
df_clean.loc[df_clean['ชั่วโมงเรียน_ต่อสัปดาห์'] < 5, 'ชั่วโมงเรียน_ต่อสัปดาห์'] = np.nan
df_clean.loc[df_clean['ชั่วโมงเรียน_ต่อสัปดาห์'] > 40, 'ชั่วโมงเรียน_ต่อสัปดาห์'] = np.nan
```

เปลี่ยนเพศที่มีคำแตกต่างออกไปให้มีคำว่า ชายและหญิง เหมือนกัน

```python
# แปลงเพศให้เป็นมาตรฐาน
gender_mapping = {
        'ชาย': 'ชาย',
        'หญิง': 'หญิง',
        'ญ': 'หญิง',
        'ช': 'ชาย',
        'M': 'ชาย',
        'F': 'หญิง',
        'Male': 'ชาย',
        'Female': 'หญิง'
    }

df_clean['เพศ'] = df_clean['เพศ'].map(gender_mapping)
```

เปลี่ยนคณะให้มีคำพื้นฐานเหมือนกัน

```python
# แปลงคณะให้เป็นมาตรฐาน
faculty_mapping = {
        'วิทยาศาสตร์': 'Science',
        'วิทย์': 'Science',
        'Science': 'Science',
        'Sci': 'Science',
        'คณะวิทยาศาสตร์': 'Science',
        'วิศวะ': 'Engineering',
        'วิศวกรรม': 'Engineering',
        'วิศวฯ': 'Engineering',
        'วิศวกรรมศาสตร์': 'Engineering',
        'Engineering': 'Engineering',
        'ศิลปศาสตร์': 'Liberal Arts',
        'ศิลป์': 'Liberal Arts',
        'Arts': 'Liberal Arts',
        'Liberal Arts': 'Liberal Arts'
    }

df_clean['คณะ'] = df_clean['คณะ'].map(faculty_mapping)
```

แทนที่ missing values ด้วย mode

```python
for col in ['เพศ', 'คณะ']:
    df_clean[col].fillna(df_clean[col].mode()[0], inplace=True)
```

แทนที่ missing values ด้วย mean

```python
for col in [ 'อายุ', 'GPA', 'ส่วนสูง_cm', 'น้ำหนัก_kg', 'ชั่วโมงเรียน_ต่อสัปดาห์']:
    df_clean[col] = df_clean[col].fillna(df_clean[col].mean())
df_clean['อายุ'] = df_clean['อายุ'].astype(np.double)
```

เปลี่ยนชื่อที่เป็น NaN ให้เป็น ไม่ระบุชื่อ

```python
df_clean['ชื่อ'] = df_clean['ชื่อ'].fillna('ไม่ระบุชื่อ')
```

บันทึก dataset ให้เป็นไฟล์ student_cleaned.csv

```python
# Save cleaned dataset
df.to_csv("student_cleaned.csv", index=False)
```
